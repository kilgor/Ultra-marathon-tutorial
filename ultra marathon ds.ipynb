```python
#with this code we can convert the distance column to km and mi and create new columns for them


def convert_distance(distance):
    distance = str(distance)
    
    try:
        if distance.endswith('mi'):
            distance_km = float(distance[:-2]) * 1.60934
            distance_mi = float(distance[:-2])
        elif distance.endswith('km'):
            distance_km = float(distance[:-2])
            distance_mi = float(distance[:-2]) / 1.60934
        else:
            distance_km = None
            distance_mi = None
    except ValueError:
        distance_km = None
        distance_mi = None
        
    return distance_km, distance_mi

df[['AD km', 'AD mi']] = df['Event distance/length'].apply(lambda x: pd.Series(convert_distance(x)))
```

```python
df.head()
```

```python
#I would like to drop NaN values except 'AD mi' and 'AD km' columns
df.dropna(subset=['AD mi', 'AD km'], inplace=True)
```

```python

#before that we will convert 'Athlete performance' column to timedelta
df['Athlete performance'] = pd.to_timedelta(df['Athlete performance'])

# Convert 'Event date' to datetime, handling errors
df['Event date'] = pd.to_datetime(df['Event date'], format='%d-%m-%Y', errors='coerce', dayfirst=True)

```

```python
#athlete club columns is unnecessary, we will drop it
df.drop(columns=['Athlete club'], inplace=True)
```
